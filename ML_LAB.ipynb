```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

Pandas: Used for data manipulation and reading CSV files.

NumPy: Used for numerical operations.

StandardScaler from scikit-learn: Standardizes the numeric columns (scaling them to have a mean of 0 and a standard deviation of 1).

PCA (Principal Component Analysis) from scikit-learn: A dimensionality reduction technique that is used to reduce the number of features in a dataset while retaining as much variability as possible.



```python
data = pd.read_csv(r"C:\Users\DELL\Desktop\family_anxiety_14_dataset.csv")
```

**Loading dataset**

```python
data.fillna(data.mean(numeric_only=True), inplace=True)
```

**Handling Missing Values:** using values in the numeric columns are handled by replacing them with the mean of the respective columns

```python
cat_features = [feature for feature in data.columns if data[feature].dtype == 'O']
numeric_features = [feature for feature in data.columns if data[feature].dtype != 'O']
```

The columns are classified into categorical columns (cat_features) and numeric columns (numeric_features) based on the data type. Categorical columns are those with a data type of 'O' (Object), while numeric columns are the rest

```python
data = pd.get_dummies(data, columns=cat_features)
```

```python
scaler = StandardScaler()
data[numeric_features] = scaler.fit_transform(data[numeric_features].values)
```

```python
pca = PCA(n_components=15)
data_pca = pca.fit_transform(data)
```

PCA is applied to reduce the dimensionality of the dataset

pca.fit_transform(data) is used to perform PCA and transform the data into the lower-dimensional space
